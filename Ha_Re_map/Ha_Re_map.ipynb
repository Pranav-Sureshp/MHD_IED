```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
class parameters :
    def __init__(self, L, B0, sigma, mu, rho, vel):
        self.L  = L
        self.B0 = B0 
        self.sigma = sigma 
        self.mu    = mu 
        self.rho   = rho 
        self.vel   = vel 
        
        self.Ha = self.L*self.B0*np.sqrt(self.sigma/self.mu)
        self.Re = vel*rho*L/mu
        self.N =  self.Ha**2/self.Re
        
```

```python
# unit 1
L = 15.7*pow(10,-3)/1.0
B = np.array([0.2, 0.85, 1.5])
V = np.array([0.2, 2.1, 4.0])
sigma = 3.23*pow(10,6)
mu = 4.3*pow(10,-4)
rho = 502.2

unit1 = []
for i in range(3):
    for j in range(3):
        case = parameters(L, B[i], sigma, mu, rho, V[j])
        unit1.append(case)
        
```

```python
# unit 2
L = 26.8*pow(10,-3)/1.0
B = np.array([0.2, 0.85, 1.5])
V = np.array([0.2, 2.1, 4.0])
sigma = 3.12*pow(10,6)
mu = 3.92*pow(10,-4)
rho = 497.6

unit2 = []
for i in range(3):
    for j in range(3):
        case = parameters(L, B[i], sigma, mu, rho, V[j])
        unit2.append(case)
        
        
```

```python
# unit 3
L = 40.0*pow(10,-3)/1.0
B = np.array([0, 0.7, 1.4])
V = np.array([0.25, 1.14, 2.02])
sigma = 3.08*pow(10,6)
mu = 4.11*pow(10,-4)
rho = 500.1

unit3 = []
for i in range(3):
    for j in range(3):
        case = parameters(L, B[i], sigma, mu, rho, V[j])
        unit3.append(case)
        
```

```python
# unit 4
L = 5.0*pow(10,-3)/1.0
B = np.array([0, 0.7, 1.4])
V = np.array([0.21, 0.74, 1.26])
sigma = 3.08*pow(10,6)
mu = 4.11*pow(10,-4)
rho = 500.1

unit4 = []
for i in range(3):
    for j in range(3):
        case = parameters(L, B[i], sigma, mu, rho, V[j])
        unit4.append(case)
        
        
```

```python
unit1_param = np.zeros((9,2))
unit2_param = np.zeros((9,2))
unit3_param = np.zeros((9,2))
unit4_param = np.zeros((9,2))

for i in range(9):
    unit1_param[i,0] = unit1[i].Re; unit1_param[i,1] = unit1[i].Ha # Re, Ha
    unit2_param[i,0] = unit2[i].Re; unit2_param[i,1] = unit2[i].Ha # Re, Ha
    unit3_param[i,0] = unit3[i].Re; unit3_param[i,1] = unit3[i].Ha # Re, Ha
    unit4_param[i,0] = unit4[i].Re; unit4_param[i,1] = unit4[i].Ha # Re, Ha
    
    
    
```

```python
# plots - analytical

Ha = np.linspace(0,5000,10)
Re1 = 200.0*Ha
Re2 = 65.0*np.sqrt(Ha)
```

```python
# Define plot space
font = {'family' : 'DeJavu Serif',
        'size'   : 20}
mpl.rc('font', **font)

fig, ax = plt.subplots(figsize=(10, 6))    

ax.plot(unit1_param[:,1], unit1_param[:,0], 'rs', markersize=10, label='Unit1')
ax.plot(unit2_param[:,1], unit2_param[:,0], 'b*', markersize=10, label='Unit2')
ax.plot(unit3_param[:,1], unit3_param[:,0], 'mv', markersize=10, label='Unit3')
ax.plot(unit4_param[:,1], unit4_param[:,0], 'go', markersize=10, label='Unit4')

ax.plot(Ha, Re1, 'k-', linewidth=1.5, label=r'$200 Ha$')
ax.plot(Ha, Re2, 'k--', linewidth=1.5, label=r'$65 Ha^{0.5}$')

plt.legend(fontsize=18,bbox_to_anchor=(1, 0.8))
plt.xlabel('Ha')
plt.ylabel('Re')
plt.xlim([-30,2500*2])
plt.ylim([-30,70000*2])

plt.savefig('./Re_Ha_plot.png',dpi=100)
```
